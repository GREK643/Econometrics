##  Test Jarque Bera

```python
K = model9.resid.kurtosis()+3
S = model9.resid.skew()
n = model9.nobs
print(K)
print(S)
```

```python
JB_statystyka_test=(n/6)*(S**2+(1/4)*(K-3)**2)
JB_statystyka_test
```

```python
JB_krytyczne = scipy.stats.chi2.sf(JB_statystyka_test, 2)
JB_krytyczne
```

Statystyka testowa wiÄ™ksza od wartoÅ›ci krytycznej odrzucamy hipotezÄ™ o rozkÅ‚adzie normlanym skÅ‚adnika losowego
